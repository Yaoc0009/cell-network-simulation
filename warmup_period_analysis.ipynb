```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
no_res_data = pd.read_csv("warmup_period_no_reservation_time.csv")[['time', 'blocked_rate', 'dropped_rate']]
start_indexes = no_res_data[no_res_data['time'] == 0].index.tolist()
```

```python
warmup_period = 300000
# Set the display
sns.set_style('whitegrid')
sns.set(rc={'figure.figsize':(20, 8)}, font_scale=1.3)

# Set 6 subplots
fig, axs = plt.subplots(nrows=2, ncols=5)
fig.tight_layout(pad=1.0)

for i in range(len(start_indexes)):
    single_data = no_res_data.iloc[start_indexes[i]:start_indexes[i] + warmup_period][['blocked_rate', 'dropped_rate']].reset_index(drop=True)
    sns.lineplot(data=single_data, ax=axs[i//5, i%5])

fig.savefig("warmup_period_analysis_no_reservation.png")
```
